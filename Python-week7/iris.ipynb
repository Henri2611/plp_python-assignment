```python
#Task 1 - Data Loading and Exploration
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Iris dataset  URL
url = "https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv"

# Load the dataset
df = pd.read_csv(url)
print("Dataset loaded successfully!")

# Explore
print("\nFirst 5 rows:")
print(df.head())

print("\nData types:")
print(df.dtypes)

print("\nMissing values per column:")
print(df.isna().sum())

```

```python
#Task 2 - Basic Data Analysis
# Basic statistics
print("\nDescriptive statistics:")
print(df.describe())

# Grouping example: average sepal_length per species
grouped = df.groupby("species")["sepal_length"].mean()
print("\nAverage Sepal Length by Species:")
print(grouped)
```

```python
#Task 3 - Data Visualization
# 1. Histogram of Sepal Length
sns.histplot(df["sepal_length"], kde=True)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# 2. Bar chart: Average petal width per species
sns.barplot(x="species", y="petal_width", data=df, estimator=np.mean)
plt.title("Average Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()

# 3. Scatter plot: Sepal Length vs Petal Length
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df)
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()

# 4. Line plot (no time variable, so simulate by index)
plt.plot(df.index, df["sepal_length"], label="Sepal Length")
plt.title("Sepal Length Trend (by Index)")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()
```
